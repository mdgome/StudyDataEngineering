**미션-미완성 문법 완성하기**

```python
class UserClass:    
    def __init(intValue=10):
        instanceVar = ''
        self.intValue = intValue
        print('난 생성자')
        
    def setVar():
        self.instanceVar = '인스턴스 변수'
        
    def getVar():
        self.instanceVar
    
    def getIntValue():
        print("클래스 내부의 getIntValue() 함수")
        return instanceVar

    def methodReturnFun():
        return getIntValue()
```

```python
instance = UserClass()
instance.setVar()

print("instanceVar 변수 값 : ", instance.getVar())
print("intValue 변수 값 : ", instance.getIntValue())    
print("intValue 변수 값 : ", instance.methodReturnFun())    
```

### 4. private 변수

- 개념 : 외부에서 호출 불가인 변수로 선언하는 방식

- 필요성 : 나이가 1살 이상이어야 할 경우 처럼 유효성 로직 실행이 필수인 경우.

- 문법 : 언더바 두개로 선언
    - 언더바 표기는 직접 호출을 비추하는 구조
    
- 권장사항 : 변수가 private인 경우 set/getXxx메소드 추가 구현 권장<br>
단, setXxx() 는 유효성 검증 코드 포함

- 명시적으로 private 변수값 수정하는 문법이 선행이 되면 private 기질을 잃음
              

```python
class People22:
    def __init__(self, name="이이", age=0):
        self.name = name
        self.__age = age   # 외부에서 절대 값 수정을 위해 호출하지 마세요 라는 의미의 표현 : private 변수
            
    def getAge(self):
        return self.__age

    # 데이터 보호 처리 기술 - 유효성 검증 코드 
    def setAge(self, new_age):
        if new_age > 0: 
            self.__age = new_age   
```

```python
p6 = People22()
print(p6.name)
print(p6.age)
```

```python
p7 = People22()
p7.age = 100
print(p7.age)
```

### 7. 메소드 재정의 (overriding/override)


**이미 존재하는 메소드의 로직 수정 방식**

> 사칙연산 식을 사용자 정의 로직의 코드로 재수정 가능

> 객체의 연산시에 사용되는 사칙 연산 관련 특수 메소드 재정의 하기 <br>
> 제공받은 특수 메소드들 재정의 or 사용자가 직접 구현한 사용자 정의 클래스 메소드를 재정의 할 수도 있음

```python
class SpecialMethod2:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __add__(self, you):
        print("__add__ 호출 확인")
        return SpecialMethod2(self.x + you.x, self.y + you.y)
    
    def __eq__(self, you):
        print('__eq__ 호출 확인')
        return self.x == you.x and self.y == you.y
    
    def __str__(self):
        print("__str__ 호출 확인")
        return '(%d, %d)' % (self.x, self.y)
```

```python
s21 = SpecialMethod2(10, 20)
s22 = SpecialMethod2(30, 40)
```

```python
v = s21 + s22    
```

```python
s21 == s22      # __eq__
```

```python
print(v)       # __str__
```

```python
print(s21 == s22)
s23 = SpecialMethod2(30, 40)
print(s22 == s23)
```
