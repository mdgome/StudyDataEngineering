```python
from pyspark.sql import SparkSession
from datetime import datetime, timedelta
import pandas as pd
import re

spark = SparkSession.builder.master('local[2]').appName('airflow log test').getOrCreate()
data = spark.read.text("MySQL_to_CSV.text") # change path
    
# value 열에서 데이터 추출
sample_logs = [item['value'] for item in data.collect()]

# 정규식 사용해 timestamp 추출 
ts_pattern = re.compile(r'\[(\d{4}-\d{2}-\d{2} \d{2}:\d{2}:\d{2},\d{3})]')

# timestamp가 존재하는 열 추출 
time_ext_log = [item for item in sample_logs if ts_pattern.match(item)]


# timestamp parsing 
Timestamp = [ts_pattern.match(item).group(1) for item in time_ext_log if ts_pattern.match(item)]


# Status Prsing List 
Status = list()

# Message Parsing List 
Message = list()

for row in time_ext_log:
    
    if "INFO" in row :
        Status.append("INFO") # Status Parsing 
        Message.append(row.split("INFO - ")[1]) # Message Parsing 
        
    elif "WARNING" in row :
        Status.append("WARNING") # Status Parsing 
        Message.append(row.split("WARNING - ")[1])  # Message Parsing 
        
    elif "ERROR" in row :
        Status.append("ERROR") # Status Parsing 
        Message.append(row.split("ERROR - ")[1])  # Message Parsing 
        
        
df = pd.DataFrame({'Timestamp' : Timestamp, 'Status':Status, 'Message':Message})  # 데이터 프레임 생성 
js = df.to_json(orient = 'records')     # json 파일 변경 
```

```python
from elasticsearch import Elasticsearch
from elasticsearch import helpers

# bulk generator function with the es required formatting for successful bulk upload. 
# the required dict is the index and source line below 
def bulk_gen(json_data):

    bulk_action = ({"_index": es_index,"_source": j,} for j in json_data if json_data is not None)
    
    for b in bulk_action:
        yield b
        print(f'processed {es_index}')

# connect to elasticsearch 
es = Elasticsearch()
# name index 
es_index = 'index name here'

# create index, add mapping
es.indices.create(index=es_index, body=data, ignore=400)

# finally, let's upload everything! 
# we've already created an index and mapping and processed the json data into bulk upload format, now we let 'er rip       

try: 
    helpers.bulk(es, bulk_gen(js))
    print('Imported resource succesfully')
except Exception as e: 
    print(f'there was an error: {e}')
```
